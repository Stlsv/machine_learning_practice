```python
with Image.open('geom.png') as img2:
  img2.load()
img2 = np.array(img2.convert("L"))

plt.imshow(img2)
plt.show()

input_data2 = np.expand_dims(img2, axis=0)

print(input_data.shape)
```

```python
conv1 = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(4, 4), stride=2)

input_tensor2 = torch.from_numpy(input_data2).to(torch.float32)
print(input_tensor2.size())

output_tensor2 = conv1(input_tensor2)
print(output_tensor2.size())

np_out2 = output_tensor2.detach().numpy()
print(np_out2.shape)

plot3d(np_out2)
```

```python
relu = nn.ReLU()

output_tensor3 = relu(output_tensor2)
print(output_tensor3.size())

np_out3 = output_tensor3.detach().numpy()
print(np_out3.shape)

plot3d(np_out3)
```

```python
tanhh = nn.Tanh()

output_tensor4 = tanhh(output_tensor2)
print(output_tensor4.size())

np_out4 = output_tensor4.detach().numpy()
print(np_out4.shape)

plot3d(np_out4)
```
